```python
import matplotlib.pyplot as plt
x=["A","B","C"]
y=[1,2,3]
plt.xlabel("language")
plt.ylabel("No")
plt.title("Matplot Library")
plt.bar(x,y,width=0.5,color="y")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=["A","B","C"]
y=[1,2,3]
plt.xlabel("language")
plt.ylabel("No")
plt.title("Matplot Library")
c=["y","b","m","g"]
plt.bar(x,y,color=c)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=["A","B","C"]
y=[1,2,3]
plt.xlabel("language")
plt.ylabel("No")
plt.title("Matplot Library")
c=["y","b","m","g"]
plt.bar(x,y,color=c,edgecolor="r",linewidth=5)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = ["A", "B", "C"]
y = [1, 2, 3]
z = [20, 30, 40]

plt.xlabel("Language")
plt.ylabel("No")
plt.title("Matplotlib Library")

# Bar plot for y values
plt.bar(x, y, width=0.4, color="r", label="abc")

# Bar plot for z values
plt.bar(x, z, width=0.4, color="y", label="abcd")

# Add legend
plt.legend()

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data for the pie chart
sizes = [40, 30, 20, 10]  # Represents the sizes (percentages) of each category
labels = ['Category A', 'Category B', 'Category C', 'Category D']  # Labels for each slice

# Create a pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%')

# Display the plot
plt.show()
```
